# Exemple

Détermination de la relation linéaire entre deux variables (constante de gravité de la terre).
Pensons aux cours de physique à l’école. Là, nous avons appris que la gravitation du
la terre accélère la vitesse de chaque objet qui tombe d'une constante g d'environ g = 9: 801 m / s 2
près de l'équateur. Cela signifie qu'un objet qui tombe augmente librement sa vitesse de 9,801 m / s 2
par seconde, il tombe.
L'école était il y a longtemps pour la plupart d'entre nous; cependant, et donc maintenant, nous vérifions cette base
constante à nouveau, à l'aide de l'analyse de régression. Selon la physique, il y a un linéaire
relation, h = gxS, où h est la hauteur à partir de laquelle un objet tombe, et S est la
temps au carré en secondes dont l'objet a besoin pour atteindre le sol.
Dans cet exemple, il n'y a qu'une seule variable d'entrée, et donc ces types de modèles sont
appelé univarié ou simple; mais les modèles de régression peuvent également être utilisés pour estimer davantage
connexions variables linéaires complexes. Dans ces cas, les modèles sont appelés multiples linéaires
modèles de régression.


Fractionner les variables comme formation et test

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, 
                                                    test_size = 0.3, random_state = 100)
```

```python
X_train
y_train
```

```python
import statsmodels.api as sm
X_train_sm = sm.add_constant(X_train)
lr = sm.OLS(y_train, X_train_sm).fit()
lr.params
lr.summary()
```


Visualiser la ligne de régression

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, 13.3750 + 0.9783*X_train, 'r')
plt.show()
```

```python
y_train_pred = lr.predict(X_train_sm)
res = (y_train - y_train_pred)
fig = plt.figure()
sns.distplot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.xlabel('y_train - y_train_pred', fontsize = 15)
plt.show()
```

```python
plt.scatter(X_train,res)
plt.show()
```

```python
X_test_sm = sm.add_constant(X_test)
y_test_pred = lr.predict(X_test_sm)
y_test_pred
```

```python
# Importing r2_square
from sklearn.metrics import r2_score

r_squared = r2_score(y_test, y_test_pred)
r_squared
```

```python
# Visualize the line on the test set
plt.scatter(X_test, y_test)
plt.plot(X_test, y_test_pred, 'r')
plt.show()
```

```python
# Splitting the data into train and test
from sklearn.model_selection import train_test_split
X_train_lm, X_test_lm, y_train_lm, y_test_lm = train_test_split(X, y, train_size = 0.7, 
                                                                test_size = 0.3, random_state = 100)
# Shape of the train set without adding column
X_train_lm.shape

# Adding additional column to the train and test data
X_train_lm = X_train_lm.values.reshape(-1,1)
X_test_lm = X_test_lm.values.reshape(-1,1)

print(X_train_lm.shape)
print(X_test_lm.shape)
```

```python
# Making Predictions of y_value
y_train_pred = lm.predict(X_train_lm)
y_test_pred = lm.predict(X_test_lm)

# Comparing the r2 value of both train and test data
print(r2_score(y_train,y_train_pred))
print(r2_score(y_test,y_test_pred))
```

# Exemple 3

Régression linéaire avec le nœud de régression Dans cet exercice, nous construisons un modèle de régression pour le Données sur le logement à Boston, Housing.data.txt, avec le nœud Regression.

1.Importez les données et spécifiez les types de variables avec le nœud Type.

```python
nabil = pd.read_csv('Datasets/housing.data.txt' , sep=",")
# Affichage de la taille du dataset (n_lignes and n_colonnes)
print("housing's shape : ", nabil.shape)
# Affichage des 10 premières lignes
nabil.head(11)
```

```python
nabil.isnull().sum()
```

2.Ajoutez un nœud de régression au flux et sélectionnez MEDV comme variable cible et toutes les autres variables comme variable contribution.

```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd  
import seaborn as sns 
%matplotlib inline
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(nabil['MEDV'], bins=30)
plt.show()
```

```python
correlation_matrix = nabil.corr().round(2)
# annot = True to print the values inside the square
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
plt.figure(figsize=(20, 5))

features = ['LSTAT', 'RM']
target = nabil['MEDV']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = nabil[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MEDV')
```

```python
X = pd.DataFrame(np.c_[nabil['LSTAT'], nabil['RM']], columns = ['LSTAT','RM'])
Y = nabil['MEDV']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score 

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
# model evaluation for training set
y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")

# model evaluation for testing set
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```
